```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

wine = pd.read_csv('example_data/wine.csv')
values = wine.values[:,1:12].astype(float)
```

```python
selection = values[::25]
distmat = ((selection[:, None] - selection[None]) ** 2).sum(axis=2)
```

```python
m = distmat.mean(axis=0)
A = -0.5 * (distmat - m[None, :] - m[:, None] + m.mean())
vals, vecs = np.linalg.eigh(A)
vecs = vecs[:, ::-1]
vals = vals[::-1]
```

```python
plt.plot(np.arange(1, 11), vals[:10])
plt.show()
```

```python
pcoa = vecs[:, :4] * vals[:4]**0.5
fig, ax = plt.subplots(figsize=(6,6))
ax.axis('equal')
ax.scatter(pcoa[:,0], pcoa[:,1], c=pcoa[:, 2], s=1)
plt.show()
```
